```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv(r"D:\\DataScience\\seaborn-data-master\\bank_data.csv")

dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
sns.kdeplot(data=dataset["installment"])
plt.show()
```

```python
sns.scatterplot(x="installment",y="int.rate",data=dataset,hue="delinq.2yrs")
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset["delinq.2yrs"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
gnb= GaussianNB()
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test)*100, gnb.score(x_train,y_train)*100
```

```python
gnb.predict([[228.22,0.1071]])
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gnb)
plt.show()
```

```python
mnb= MultinomialNB()
mnb.fit(x_train,y_train)
```

```python
mnb.score(x_train,y_train)*100, mnb.score(x_test,y_test)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mnb)
plt.show()
```

```python
bnb= BernoulliNB()
bnb.fit(x_train,y_train)
```

```python
bnb.score(x_train,y_train)*100, bnb.score(x_test,y_test)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=bnb)
plt.show()
```
