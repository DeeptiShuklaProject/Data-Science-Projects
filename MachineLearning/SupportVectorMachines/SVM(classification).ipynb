```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv(r"D:\\DataScience\\seaborn-data-master\\bank_data.csv")

dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
sns.scatterplot(x="installment",y="int.rate",data=dataset,hue="delinq.2yrs")
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset["delinq.2yrs"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
sv=SVC(kernel="linear")
sv.fit(x_train,y_train)
```

```python
sv.score(x_train,y_train)*100
```

```python
sv.score(x_test,y_test)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=sv)
```

```python
sv=SVC(kernel="poly")     #Can use Linear,poly,rbf,sigmoid,precomputed
sv.fit(x_train,y_train)
```
