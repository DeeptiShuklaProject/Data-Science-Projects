```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r"D:\\DataScience\\seaborn-data-master\\loan_data.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset["installment"].fillna(dataset["installment"].mean(),inplace=True)
```

```python
dataset.isnull().sum()
```

```python
sns.displot(dataset["installment"])
plt.show()
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
ss.fit(dataset[["installment"]])
```

```python
ss.transform(dataset[["installment"]])
```

```python
dataset["installment_ss"]=pd.DataFrame(ss.transform(dataset[["installment"]]),columns=["x"])
```

```python
dataset.head(3)
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("After")
sns.distplot(dataset["installment_ss"])

plt.subplot(1,2,2)
plt.title("Before")
sns.distplot(dataset["installment"])

plt.show()
```
