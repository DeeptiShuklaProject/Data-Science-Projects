```python
#Feature Scaling Technique
#2.Normalization:It is also called min-mac scaler.it is a scaling technique in which values are shifted and rescaled so that they end up ranging between 0 and 1.it is also known as Min-Max scaling.
# Xnew=(Xi-min(X))/max(X)-min(X)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r"D:\\DataScience\\seaborn-data-master\\titanic.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
sns.distplot(dataset["fare"])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms=MinMaxScaler()
ms.fit(dataset[["fare"]])
```

```python
dataset["fare_min"]=ms.transform(dataset[["fare"]])
dataset.head(3)
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("After")
sns.distplot(dataset["fare_min"])

plt.subplot(1,2,2)
plt.title("Before")
sns.distplot(dataset["fare"])

plt.show()
```
