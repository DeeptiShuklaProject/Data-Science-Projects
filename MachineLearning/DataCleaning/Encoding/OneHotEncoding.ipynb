```python
dataset = pd.read_csv(r"D:\\DataScience\\seaborn-data-master\\tips.csv")
dataset.head(4)
```

```python
en_data=dataset[["Gender","smoker"]]
```

```python
pd.get_dummies(en_data).info()
```
