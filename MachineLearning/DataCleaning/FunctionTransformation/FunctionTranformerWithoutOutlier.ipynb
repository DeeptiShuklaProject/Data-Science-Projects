```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv(r"D:\\DataScience\\seaborn-data-master\\titanic.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
sns.distplot(dataset["fare"])
plt.show()
```

```python
q1=dataset["fare"].quantile(0.25)
q3=dataset["fare"].quantile(0.75)
iqr=q3-q1
```

```python
min_r=q1-(1.5*iqr)
max_r=q3+(1.5*iqr)
min_r,max_r
```

```python
dataset=dataset[dataset["fare"]<=max_r]
```

```python
sns.distplot(dataset["fare"])
plt.show()
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
ft = FunctionTransformer(func=np.log1p)
```

```python
ft.fit(dataset[["fare"]])
```

```python
dataset["fare_tf"]=ft.transform(dataset[["fare"]])
```

```python
plt.figure(figsize=(10,7))
plt.subplot(1,2,1)
sns.distplot(dataset["fare"])
plt.title("Before")

plt.subplot(1,2,2)
sns.distplot(dataset["fare_tf"])
plt.title("After")
plt.show()
```
