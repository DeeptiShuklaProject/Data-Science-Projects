```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv(r"D:\\DataScience\\seaborn-data-master\\titanic.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
sns.distplot(dataset["fare"])
plt.show()
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
ft1=FunctionTransformer(func=lambda x: x**2)
ft1.fit(dataset[["fare"]])
dataset["fare_tf"]=ft1.transform(dataset[["fare"]])
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.distplot(dataset["fare"])
plt.title("Before")

plt.subplot(1,2,2)
sns.distplot(dataset["fare_tf"])
plt.title("After")
plt.show()
```
