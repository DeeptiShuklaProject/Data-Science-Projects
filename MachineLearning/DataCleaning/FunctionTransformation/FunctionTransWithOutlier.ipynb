```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv(r"D:\\DataScience\\seaborn-data-master\\titanic.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
sns.distplot(dataset["fare"])
plt.show()
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
ft = FunctionTransformer(func=np.log1p)
```

```python
ft.fit(dataset[["fare"]])
```

```python
dataset["fare_tf"]=ft.transform(dataset[["fare"]])
```

```python
plt.figure(figsize=(10,7))
plt.subplot(1,2,1)
sns.distplot(dataset["fare"])
plt.title("Before")

plt.subplot(1,2,2)
sns.distplot(dataset["fare_tf"])
plt.title("After")
plt.show()
```
